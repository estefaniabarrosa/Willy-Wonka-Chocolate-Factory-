```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Mysql@123",
    database="wonka_choc_factory"
)

# SQL query
query = """
SELECT
    l.region,
    SUM(od.sales) AS regional_sales
FROM order_details od
JOIN orders o
ON od.order_id = o.order_id
JOIN locations l
ON o.location_id = l.location_id
GROUP BY l.region
ORDER BY regional_sales DESC;
"""

# Load dataframe
df = pd.read_sql(query, conn)

# Create chart
plt.figure(figsize=(10,6))

sns.barplot(
    data=df,
    x='region',
    y='regional_sales'
)

plt.title("Regional Sales Performance")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

conn.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Mysql@123",
    database="wonka_choc_factory"
)

# SQL query
query = """
SELECT
    f.factory_name,
    SUM(od.gross_profit) AS total_profit
FROM order_details od
JOIN products p
ON od.product_id = p.product_id
JOIN factories f
ON p.factory_id = f.factory_id
GROUP BY f.factory_name
ORDER BY total_profit DESC;
"""

# Load SQL results into dataframe
df = pd.read_sql(query, conn)

# Create visualization
plt.figure(figsize=(10,6))

sns.barplot(
    data=df,
    x='factory_name',
    y='total_profit'
)

plt.title("Factory Profitability Analysis")
plt.xlabel("Factory")
plt.ylabel("Total Profit")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

conn.close()
```

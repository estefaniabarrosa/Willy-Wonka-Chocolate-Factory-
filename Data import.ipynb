```python
import pandas as pd
import mysql.connector

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Mysql@123",
    database="wonka_choc_factory"
)

cursor = conn.cursor()

# Read CSV
df = pd.read_csv(r"C:\Users\durga\Downloads\customers.csv")

# Remove duplicates inside CSV
df = df.drop_duplicates(subset=['customer_id'])

# Insert rows safely
for _, row in df.iterrows():

    cursor.execute(
        """
        INSERT IGNORE INTO customers (customer_id)
        VALUES (%s)
        """,
        (int(row['customer_id']),)
    )

# Commit changes
conn.commit()

print("Customers imported successfully!")

# Verify inserted rows
cursor.execute("SELECT COUNT(*) FROM customers")
print("Total rows:", cursor.fetchone()[0])

cursor.close()
conn.close()
```

```python
import pandas as pd
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Mysql@123",
    database="wonka_choc_factory"
)

cursor = conn.cursor()

df = pd.read_csv(r"C:\Users\durga\Downloads\locations.csv")

df = df.drop_duplicates(subset=['location_id'])

for _, row in df.iterrows():

    cursor.execute(
        """
        INSERT IGNORE INTO locations (
            location_id,
            country_region,
            region,
            state_province,
            city,
            postal_code,
            latitude,
            longitude
        )
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
        """,
        (
            int(row['location_id']),
            row['country_region'],
            row['region'],
            row['state_province'],
            row['city'],
            row['postal_code'],
            float(row['latitude']),
            float(row['longitude'])
        )
    )

conn.commit()

cursor.execute("SELECT COUNT(*) FROM locations")
print("Locations rows:", cursor.fetchone()[0])

cursor.close()
conn.close()
```
